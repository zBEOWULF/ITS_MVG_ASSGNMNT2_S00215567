```python
# Source: ADAPTED FROM https://www.life2coding.com/extract-frame-video-file-using-opencv-python/
import cv2
import os
import numpy as np 
import matplotlib.pylab as plt

cap = cv2.VideoCapture(os.path.expanduser('~/Desktop/Speed Test.mp4'))
count=int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
cap.release()
print('Total no of frames: ',count)


def extractFrames(pathIn, pathOut):
    os.mkdir(pathOut)

    cap = cv2.VideoCapture(pathIn)
    count = 0

    while (cap.isOpened()):

        # Capture frame-by-frame
        ret, frame = cap.read()

        if ret == True:
            print('Read %d frame: ' % count, ret)
            cv2.imwrite(os.path.join(pathOut, "frame{:d}.jpg".format(count)), frame)  # save frame as JPEG file
            count += 1
        else:
            break

    fps = cap.get(cv2.CAP_PROP_FPS)
    print("Frames per second using video.get(cv2.CAP_PROP_FPS) : {0}".format(fps))
        
    # When everything done, release the capture
    cap.release()
    cv2.destroyAllWindows()

def main():
    extractFrames(os.path.expanduser('~/Desktop/Speed Test.mp4'), 'SPEED_R')

# https://www.freecodecamp.org/news/if-name-main-python-example/
if __name__=="__main__":
    main()
```
